```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
from scipy.stats import expon
from scipy.stats import gamma
from scipy.stats import beta
from scipy.stats import geom
```

# Distributions in Statistics

Probability distributions are essential for describing random variables and probabilistic phenomena. They are used to model random variables and analyze data. Distributions allow us to assess the probability of specific events and make statistical inferences. For instance, the normal distribution is often used to model natural phenomena, while the Poisson distribution is used to model the number of events occurring in a given time or space interval.
Furthermore, distributions can be employed to test hypotheses and make decisions based on statistical data. For example, when testing whether the mean of a sample is statistically significant, the t-distribution can be used.
In general, probability distributions are a vital tool for data analysis and decision-making based on statistical information. They enable the description of random variables and probabilistic phenomena, as well as the drawing of statistical inferences from such data.

# Poisson Distribution

https://programmathically.com/the-poisson-distribution/

The Poisson distribution is useful in any scenario where you want to model the occurrence of independent events happening at a certain rate.

Let’s say you are looking at an intersection and expect cars to take a left turn at a rate of 5 per minute. You could calculate the probability of observing 3 cars turning left by using the Poisson distribution.

```python
lam = 5
k = 3

poisson_dist = poisson(mu=lam)

prob = poisson_dist.pmf(k)

x = np.arange(0, 15)


plt.bar(x, poisson_dist.pmf(x), label=f'Poisson Distribution (λ={lam})')
plt.xlabel('Number of Cars')
plt.ylabel('Probability')
plt.axvline(k, color='r', linestyle='--', label=f'k={k} (Probability={prob:.4f})')
plt.legend()
plt.show()
```

```python
print(prob)
```

The probability of obesrving 3 cars turning left at a intersection is equivalent to 14%

# Exponential Distribution

https://programmathically.com/exponential-distribution/

The exponential distribution models the time interval between continuously occurring, independent events. Exponential distribution models the wait time between events in a Poisson process.

Let’s assume that this function models the wait time for a bus arriving once per hour at a bus stop. If we wanted to calculate the probability that the bus arrives two hours or later, we could plug the following values into the function.

```python
exp_dist = expon(scale=1)
prob = 1 - exp_dist.cdf(2)

x = np.linspace(0, 10, 100)
y = exp_dist.pdf(x)
plt.plot(x, y, label='Exponential Distribution PDF')

x_fill = np.linspace(2, 10, 100)
y_fill = exp_dist.pdf(x_fill)
plt.fill_between(x_fill, y_fill, color='orange', label='Probability of arriving 2 hours or later')

plt.xlabel('Time (hours)')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
print(prob)
```

Probability of observing a bus arriving at or after 2 hours of waiting is 13%

# Gamma Distribution

https://programmathically.com/gamma-distribution/

Assume you are waiting at a bus station. Busses are passing by at a rate of two per hour. You want to know the probability that you have to wait less than two hours until at least three busses have passed. The distribution of wait times for busses looks like this.

```python
rate = 2
a = 3

g_dist = gamma(a=a, scale=(1/rate))

prob = g_dist.cdf(2)

x = np.linspace(0, 10, 100)

plt.plot(x, g_dist.pdf(x), label=f'Gamma Distribution (α={a}, λ={rate})')
plt.xlabel('Time (hours)')
plt.ylabel('Probability')
plt.axvline(2, color='r', linestyle='--', label=f'Probability={prob:.4f}')
plt.legend()
plt.show()
```

```python
print(prob)
```

The probability that 3 busses have passed by after two hours is 76% when the average arrival rate is 2 busses per hour.

# Beta Distribution

https://programmathically.com/the-beta-distribution/

The beta distribution models a distribution of probabilities. If we don’t know the probability of the outcome of an event, we can use the beta distribution to model the distribution of probabilities given the prior information.

Let’s say you have 20 customers. 10 of them buy and 10 leave without buying. The average probability of making a purchase or conversion rate is now 50% (0.5). But 20 people isn’t a very reliable sample size. Your real probability of converting customers could be lower or higher.

```python
a = 10
b = 10

beta_dist = beta(a=a, b=b)

x = np.linspace(0, 1, 100)

plt.plot(x, beta_dist.pdf(x), label=f'Conversion Rate: 50%')
plt.xlabel('Conversion Rate')
plt.ylabel('Probability Density')
plt.axvline(0.5, color='r', linestyle='--', label='Current Conversion Rate')
plt.legend()
plt.show()
```

# Geometric Distribution

https://programmathically.com/geometric-distribution-and-geometric-random-variables/

The geometric distribution describes the probability of the number of failures before a successful outcome in a Bernoulli trial.

Suppose you are a recruiter and you need to find a suitable candidate to fill an IT job. Your chance that a suitable candidate will be interested is 15%. Let’s calculate the probability of finding a suitable candidate on the third attempt.

```python
p = 0.15
x = 3

ge_dist = geom(p=p)

prob = ge_dist.cdf(x)

x = np.arange(0, 10)

plt.bar(x, ge_dist.cdf(x), label=f'Geometric Distribution (p={p})')
plt.xlabel('Trials No.')
plt.ylabel('Probability')
plt.axvline(3, color='r', linestyle='--', label=f'Probability={prob:.4f}')
plt.legend()
plt.show()
```
